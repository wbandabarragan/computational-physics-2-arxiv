# FFT method: One-way wave equation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Wave speed
c = 2 

# Length of domain
L = 20

# Number of points 
N = 100

# Step size
delta_x = L/N

# X-axis vector

x = np.arange(-L/2, +L/2, delta_x)

print(x.shape)
```

```python
# Initial condition

rho_0 = 1/np.cosh(x)
```

```python
# Plotting

plt.figure(figsize=(3,2))

plt.plot(x, rho_0)

plt.show()
```

## Part 1 : FFT method

### Simulation in Fourier (k) domain

```python
# Wavemunbers - spatial frequencies:

k_numbers = 2*np.pi*np.fft.fftfreq(len(x), d=delta_x)

#print(k_numbers.shape)

#print(k_numbers)
```

```python
# FFT of the initial condition

rho_0_hat = np.fft.fft(rho_0)

#print(rho_0_hat)
```

```python
# Norm of Fourier transform

ampli_rho_0_hat = np.abs(rho_0_hat)/(rho_0_hat.size//2)
```

```python
# Plotting

plt.figure(figsize = (4, 3))
plt.stem(k_numbers, ampli_rho_0_hat, 'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Frequency')
plt.ylabel('Fourier Amplitude')

plt.show()
```

### Concatenate real and complex arrays

ODE integrators don't like complex numbers.

```python
rho_0_conc = np.concatenate((rho_0_hat.real, rho_0_hat.imag))
print(rho_0_hat.shape, rho_0_conc.shape)
```

```python
#print(rho_0_conc)
```

### Time discretisation:

```python
# Time step
t_step = 0.025

m = 100

t = np.arange(0, m*t_step, t_step)

print(t.shape)
```

### Construct ODE (RHS of ODE)

```python
# Function to get RHS

def RHS_ODE(rho_0_conc, t, k_numbers, c):
    
    u_tilde = rho_0_conc[:N] + (1j)*rho_0_conc[N:]
    
    rhs_u_tilde = -(1j)*(c)*(k_numbers)*u_tilde
    
    rhs_ode = np.concatenate((rhs_u_tilde.real, rhs_u_tilde.imag))
    
    return rhs_ode
```

### k ODEs: solution

```python
import scipy.integrate as sint
```

```python
# Call ODE integrator

solution = sint.odeint(RHS_ODE, rho_0_conc,\
                       t, args = (k_numbers, c))

print(solution.shape)
```

### Reconstruct Complex solution:

```python
rho_solution = solution[:, :N] + (1j)*solution[:, N:]

print(rho_solution.shape)
```

### Inverse Fourier transform of each rho_solution

```python
# For loop with k as index

inv_rho_solution = np.zeros(rho_solution.shape,\
                            dtype = complex)

for k in range(len(t)):
    
    inv_rho_solution[k, :] = np.fft.ifft(rho_solution[k, :])
    
print(inv_rho_solution.shape)
```

### Plotting the solution:

```python
# Add colour

R = np.linspace(1, 0, m)
B = np.linspace(0, 1, m)
G = 0

plt.figure(figsize= (10, 6))

for j in range(0, m, 5):
    plt.plot(x, inv_rho_solution[j, :], color = [R[j], G, B[j]])

plt.xlabel("position [m]")
plt.ylabel("density")

plt.show()
```

## Part 2 : FFT method

### Simulation in Real spatial (x) domain

```python
# Function to get RHS

def RHS_ODE_spatial(rho_0, t, k_numbers, c):
    
    # Initial condition to Fourier space
    rho_hat_0 = np.fft.fft(rho_0)
    
    # Factor in front of u_hat in Fourier space
    rhs_u_tilde = (1j)*(k_numbers)*rho_hat_0
    
    # Inverse FFT of the RHS
    rhs_u = np.fft.ifft(rhs_u_tilde).real
    
    # Back in Real Space
    rhs_ode = -c*rhs_u
    
    return rhs_ode
```

```python
A = RHS_ODE_spatial(rho_0, t, k_numbers, c)

print(A.shape)
```

### ODE Integration in real space:

```python
# Call ODE integrator

solution_spatial = sint.odeint(RHS_ODE_spatial, rho_0,\
                       t, args = (k_numbers, c))

print(solution_spatial.shape)
```

```python
# Add colour

R = np.linspace(1, 0, m)
B = np.linspace(0, 1, m)
G = 0

plt.figure(figsize= (10, 6))

for j in range(0, m, 5):
    plt.plot(x, solution_spatial[j, :], color = [R[j], G, B[j]])

plt.xlabel("position [m]")
plt.ylabel("density")

plt.show()
```
