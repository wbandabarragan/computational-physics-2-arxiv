### Heat equation: metal heated in the middle

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# This is for 3D visualisation
from mpl_toolkits.mplot3d import Axes3D
```

### Grid creation:

```python
# Steps

k = 0.025 # Time step
h = 0.025 # Spatial step
c_2 = 1.

r_factor = c_2*k/h**2
print(r_factor)

# Limits
a = 1.0
b = 0.1

# Axes
t = np.arange(0, b + k , k).round(3)
x = np.arange(0, a + h , h)

print(x.shape)
```

```python
# Boundary Conditions

bcs = [0, 0] # T = 0 at x=0, and x=a

# Initial conditions

ics = np.sin(np.pi*x) # this is the initial T at t=0
```

```python
# Matrix T

n = len(x)
m = len(t)
print(n, m)

T = np.zeros((n, m))

#print(T)
```

### Fill up the matrix elements:

```python
# Fill up the boundry conditions

T[0 , :] = bcs[0]
T[-1, :] = bcs[1]

# Fill up the initial conditions
T[: , 0] = ics

print(T.round(3).T)
```

```python
# outler loop (time loop)

for j in range(1, m):
    # inner loop (spatial loop)
    for i in range(1, n - 1):
        
        T[i, j] = (1 - 2*r_factor)*T[i, j-1] + r_factor*(T[i-1, j-1] + T[i+1, j-1])
        
```

```python
print(T.round(3).T)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x, T[:, 0], label = 't=0')

plt.plot(x, T[:, 1], label = 't=0.025')

plt.plot(x, T[:, 2], label = 't=0.050')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))

for j in range(m):
    plt.plot(x, T[:, j])


plt.xlim(0,2)

plt.xlabel("position (x)")
plt.ylabel("temperature(K)")
    
plt.legend([f'time = {time_value}' for time_value in t], loc=1)
plt.show()
```

# Max. temperature vs. time.

```python
plt.figure(figsize=(10,6))


plt.plot(t, T[int(x.shape[0])//2 , :])


plt.xlim(0,0.2)
plt.ylim(0,1)

plt.xlabel("time (s)")
plt.ylabel("temperature(K)")
    
plt.legend([f'time = {time_value}' for time_value in t], loc=1)
plt.show()
```

## Temperature surface

```python
# Create 2D grid

x_grid, t_grid = np.meshgrid(x, t)

print(x_grid.shape, t_grid.shape, T.T.shape)
```

```python
plt.figure(figsize=(10,6))

Z1 = plt.pcolor(x_grid, t_grid, T.T, cmap = 'magma')

plt.xlabel("position")

plt.ylabel("time")

plt.title("temperature")

plt.colorbar(Z1)

plt.show()
```

```python
plt.figure(figsize=(12,8))

ax  = plt.axes(projection='3d')

ax.contour3D(x_grid, t_grid, T.T, 500, cmap = 'magma')

plt.xlabel("position")
plt.ylabel("time")

plt.title("temperature")

plt.colorbar(Z1, ax=ax)

plt.show()
```
